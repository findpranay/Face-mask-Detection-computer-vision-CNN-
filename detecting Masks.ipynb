```python
from keras.models import load_model
import cv2
import numpy as np
model = load_model('model.model')
```

```python

face_clsfr=cv2.CascadeClassifier('C:/Users/Yaniv/AppData/Local/Packages/PythonSoftwareFoundation.Python.3.10_qbz5n2kfra8p0/LocalCache/local-packages/Python310/site-packages/cv2/data/haarcascade_frontalface_default.xml')

source=cv2.VideoCapture('C:/Users/Yaniv/Desktop/mask.mp4')

labels_dict={0:'NO MASK',1:'MASK'}
color_dict={0:(0,0,255),1:(0,255,0)}
```

```python
def rescaleFrame(frame,scale=0.4):
    width=int(frame.shape[1]*scale)
    height=int(frame.shape[0]*scale)
    dimensions=(width,height)
    return cv2.resize(frame,dimensions,interpolation=cv2.INTER_AREA)

```

```python
while(True):

    ret,img=source.read()
    i=rescaleFrame(img)
    gray=cv2.cvtColor(i,cv2.COLOR_BGR2GRAY)
    faces=face_clsfr.detectMultiScale(gray,1.3,5)  

    for (x,y,w,h) in faces:
    
        face_img=gray[y:y+w,x:x+w]
        resized=cv2.resize(face_img,(100,100))
        normalized=resized/255.0
        reshaped=np.reshape(normalized,(1,100,100,1))
        result=model.predict(reshaped)
        
        label=np.argmax(result,axis=1)[0]
        print(label)
      
        cv2.rectangle(i,(x,y),(x+w,y+h),color_dict[label],2)
        cv2.rectangle(i,(x,y-40),(x+w,y),color_dict[label],-1)
        cv2.putText(i, labels_dict[label], (x, y-10),cv2.FONT_HERSHEY_SIMPLEX,0.8,(255,255,255),2)
        
    cv2.imshow('LIVE',i)
    key=cv2.waitKey(1)
    
    if(key==27):
        break
        
cv2.destroyAllWindows()
source.release()
```
